```python
# path = r'Z:\CIS-PD Study\Scores'
path = '/Volumes/RT&O/CIS-PD Study/Scores/' #Mac
MDSPath = os.path.join(path,'MDS-UPDRS.xls')
MotorPath = os.path.join(path,'MotorTasks.xls')

MDSdf = pd.read_excel(MDSPath)
Motordf = pd.read_excel(MotorPath)

MDSdf2 = MDSdf.iloc[:,9:42]
Motordf2 = Motordf.iloc[:,16:151]
colNames = MDSdf2.columns.values

plt.figure(num=None, figsize=(30, 30), dpi=80, facecolor='w', edgecolor='k')

# Plotting MDS Histograms
k=0
for i in range(6):
    for j in range(6):
        list1 = MDSdf2.iloc[:,k].tolist()
        y = str(list1)
        z = [z for z in list1 if str(z) != 'nan']
        bins = np.arange(0,4,.5)
        
        ax = plt.subplot2grid((6,6), (i,j))
        ax.hist(z, bins=bins, align='left')
        ax.set_title(colNames[k])
        k=k+1
        if k > 32:
            break;
```
